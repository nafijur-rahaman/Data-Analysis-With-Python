### Line plot using seaborn

```python
import pandas as pd
import seaborn as sns

student  = pd.read_csv("./sns_data (1).csv")
student.head()
```

```python
#axes level

sns.lineplot(data = student, x = "week", y = "attendance_rate", errorbar=None)
```

```python
#categorical 
sns.lineplot(data = student, x = "week", y = "attendance_rate", errorbar=None,hue="gender")
```

```python
#figure level

sns.relplot(kind="line", data = student, x = "week", y = "attendance_rate", errorbar=None,hue = "gender")
```

```python
sns.relplot(kind="line", data = student, x = "week", y = "attendance_rate", errorbar=None,hue = "class_level")
```

### Scatter Plot

```python

sns.scatterplot(data = student, x = "study_hours", y = "test_score",hue="gender")
```

```python
sns.scatterplot(data = student, x = "study_hours", y = "test_score",hue="gender",size="Tshirt_size")
```

```python
# figure level scatter plot


sns.relplot(kind='scatter', data = student , x = 'study_hours' , y ='test_score',hue='gender',style='subject',size='Tshirt_size') 
```

```python
#load built in dataset

tips_data = sns.load_dataset("tips")
tips_data.head()
```

```python
sns.relplot(kind = "scatter", data = tips_data , x = "total_bill" , y = "tip" , hue = "sex",size="size",style="time") 
```

### facet plot

```python
sns.relplot( kind = "scatter", data = student, x = "study_hours", y = "test_score", col="gender")
```

```python
# week based facet plot

sns.relplot(kind='scatter', data = student, x = 'study_hours' , y ='test_score',col='week',col_wrap=3) 
```

### histogram using seaborn

```python
#axes level histogram

sns.histplot(data = student, x = "test_score", bins=10, hue="gender", element="step")
```

```python
#figure level

sns.displot(data = student, x = "test_score", bins=10, hue="gender", element="step")
```

```python
sns.displot(kind='hist' , data = student , x ='attendance_rate', col='gender') 
```

### KDE plot using Seaborn

```python
#axes level

sns.kdeplot(data = student, x = "attendance_rate")
```

```python
sns.kdeplot(data = tips_data, x = "tip", hue="sex", fill=True)
```

```python
#figure leve

sns.displot(kind='kde', data = tips_data, x = "tip", col="sex")
```

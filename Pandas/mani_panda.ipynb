##String Manipulation in Pandas

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)

df
```

```python
#case sensitive
df.loc[df["City"].str.contains("New")]
```

```python
#without case sensitive
df.loc[df["City"].str.contains("new",case=False)]
```

```python
# find the name start with Los
# ^ → match at start

df.loc[df["City"].str.contains(r"^Los")]  
```

```python
#end with rk
# $ → match at end
df.loc[df["City"].str.contains(r"rk$")]
```

```python
#name start with a vowel
df.loc[df["Name"].str.contains(r"^[AEIOU]")]
```

```python
# string contains new or los

df.loc[df["City"].str.contains(r"New|Los")]
```

##Adding New Column

```python
df = pd.read_csv('student_data.csv')
df.tail()
```

```python

#adding a column with a constant value also for all rows
df["Country"] = "Bangladesh"
df.head()
```

```python
df["Total_Marks"] = df["Data Structure Marks"]+df["Algorithm Marks"]+df["Python Marks"]
df.head()
```

```python
import numpy as np

df["A+ in ds"] = np.where(df["Data Structure Marks"]>=80,'A+','A')
df.head()
```

```python
df["First Name"] = df["FullName"].str.split(" ").str[0]
df.head()
```

```python
#make a new csv file with new data
df.to_csv("new_student.csv")
```

### Basic

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 7, 3, 9]
plt.title("basic plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.grid(True)
plt.plot(x, y)
plt.show()
```

### 2D Line Plot From Pandas Data

```python
import pandas as pd

df = pd.read_csv("enrollment_data.csv")
```

```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'] ,df['Programming'],label = 'Programming')
plt.plot(df['Year'] ,df['Digital Marketing'], label= 'Digital Marketing')

plt.title("Student Enrollment over 8 years")
plt.legend()
```

### visualization



```python
plt.xlabel('Years') 
plt.ylabel('Count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df["Year"],df["Programming"],label = "Programming",color = "red", linewidth = 3,linestyle="dashed",marker='o',markersize=7)
plt.plot(df["Year"],df["Digital Marketing"], label = "Digital Marketing",linewidth = 3)

plt.legend()
```

### basic plotly

```python
import plotly.express as px
import pandas as pd
fig = px.scatter(x=[1,2,3,4,5], y=[10,11,12,13,14])
fig.show()
```

```python
student = pd.read_csv("./student_dataset_complete.csv")
student.head()
student_marks = student[['Marks','study_hours','attendance_rate','gender']]
```

### Line plot using plotly

```python
fig = px.line(student_marks, x='study_hours', y='Marks', title='Study Hours vs Marks')
fig.show()
```

### histogram using plotly

```python
student_marks.head()

fig = px.histogram(student_marks, x="attendance_rate")
fig.show()
```
